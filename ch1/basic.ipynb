| 코드 | 설명 |
| --- | --- |
| `%s` | 문자열(string) |
| `%d` | 정수(integer) |
| `%c` | 문자 1개(character) |
| `%f` | 부동 소수(floating-point) |
| `%o` | 8진수 |
| `%x` | 16진수 |
| `%%` | % 문자 자체 |

- format() 함수 사용
- f 문자열 포매팅(파이썬 3.6버전부터 지원)

```python
# hi 문자열 출력 시 10개인 공간 사용
# 오른쪽 정렬

print("%10s" % "hi")

#왼쪽 정렬
print("%-10sjane" % "hi")

# 0은 길이와 상관 하지 않겠다는 의미
# 소수점 네 번째 자리까지만
print("%0.4f" % 3.42134234)
# 전체 길이 10개인 문자열 곻간에서 오른쪽으로 정렬
print("%10.4f" % 3.42134234)


```

```python
name = "홍길동"
age = 30

# 첫번째 방법
print ("나의 이름은 %s 이고, 나이는 %d세 입니다." %(name, age))

#두번째 방법
print ("나의 이름은 {} 이고, 나이는 {}세 입니다.".format(name, age))

# 세번째 방법
print (f"나의 이름은 {name} 이고, 나이는 {age}세 입니다.")
print (f"나의 이름은 {name} 이고, 나이는 {age+1}세 입니다.")
```

### 6. 집합 자료형
- 집합에 관련된 것을 쉽게 처리하기 위한 자료형
- 중복 제거 / 순서가 없음
- set()

```python
s1 = set()
s2 = set([1,2,3,4,5,4])
s3 = set([1,2,3,4,5,6])
```

```python
print(s1)
print(s2)
print(s3)
```

```python
# 인덱싱 불가
# s2[0]

# 인덱싱, 슬라이싱을 하고 싶다면 가능한 타입으로 변환
# 리스트 : list(), 튜플 tuple()

s1 = "안녕하세요"
print(type(s1)) # str = 문자열
print(type(3))
print(type(list3))
print(type(t4))
print(type(dict2))
print(type(s2)) # set = 집합

# set 자료형 => 새로운 list로 생성
print(list(s2)[2])
print(tuple(s3))
print(s2)
```

```python
# set() : list, tuple set 구조로 변환시 사용

print(set(list4))
print(set(s1))
```

```python
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])

# 교집합
print(s1 & s2)
print(s1.intersection(s2))

# 합집합
print(s1 | s2)
print(s1.union(s2))

# 차집합
print(s1 - s2)
print(s1.difference(s2))
```

```python
# 값 1개 추가하기
s1.add(8)
print(s1)

# 값 여러개 추가하기
s1.update([9,10])
print(s1)
```

```python
# 값 제거하기
s1.remove(3)
print(s1)
```
