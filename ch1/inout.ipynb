```python
### 표준 입출력
- 표준 입력 : 키보드 => input()
- 표준 출력 : 모니터 => print()
```

```python
num = input("숫자 입력")
# type(num)
# 사용자에게 숫자를 입력 받은 후 +10한 결과 출력

if num.isdigit():
    num = int(num) + 10
    print(num)
else:
    print("숫자를 입력하세요")
```

```python
# 사용자로부터 입력 받은 숫자가 4가 아니라면 반복

num = int(input("숫자 입력"))
while num != 4:
    num = int(input("숫자 입력"))
    print(num)
```

```python
# money = 300, coffee = 10
money,coffee = 300,10

#break : 반복문 강제로 빠져나가기
while money:
    print('커피 주기')
    coffee = coffee - 1
    print(f"남은 커피 개수 {coffee}")

    if coffee == 0:
        print('커피 판매 종료')
        break
```

```python
# 구구단
# 3단 출력

i = 1
while i < 10:
    print(f"{3} * {i} = {3*i}")
    i = i + 1
```

```python
# 사용자가 원하는 구구단 출력
num = int(input())

i = 1
while i < 10:
    print(f"{num} * {i} = {num*i}")
    i = i + 1
```

```python
# 1~100사이의 숫자 중에서 3의 배수만을 더한 후 결과 값 출력
i = 1
hap = 0
while i <= 100:
    if i % 3 == 0:
        hap = hap + i
    i = i + 1

print(hap)
```

```python
# 1~100사이의 숫자 중에서 3의 배수(단, 9의 배수는 빼기)만을 더한 후 결과 값 출력
i = 1
hap = 0
while i <= 100:
    if i % 3 == 0 and i % 9 != 0:
        hap = hap + i    
    i = i + 1

print(hap)
```

```python
# 1~100사이의 숫자 중에서 3의 배수와 5의 배수를 더한 후 결과 값 출력
i = 1
hap = 0
while i <= 100:
    if i % 3 == 0 or i % 5 == 0:
        hap = hap + i    
  
    i = i + 1

print(hap)
```

```python
# continue : 반복문의 처음으로 다시 돌아가기

i = 0
while i < 10:
    i = i + 1
    if i % 2 == 0: continue
    print(i)
```
