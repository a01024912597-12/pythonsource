'숫자가 짝수면 짝수라고 출력, 아니면 홀수라고 출력하기'

```python
num = 2

# if 문 : bool 자료형
if num % 2 == 0:
    print("짝수")
else:
    print("홀수")
```

```python
# 점수가 90 이상이면 A, 80 이상이면 B, 70이상이면 C, 나머지는 F

score = 99
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else: 
    print("F")
```

```python
# a = 12
# b = 6
# c = 19
a, b, c = 12, 6, 19

# 가장 큰 숫자 출력
if a > b and a > c:
    print(a)
elif b > a and b > c:
    print(b)
else :
    print(c)
```

```python
# 가장 큰 수
max = a
if max < b :
    max = b
if max < c :
    max = c
    print("가장 큰 수 :", max)
```

```python
# 가장 작은 수
min = a
if min > b:
    min = b
if min > c:
    min = c

print("가장 작은 수 :", min)
```

```python
fruits = ['딸기', '사과', '바나나', '배', '수박']

# fruits 리스트에 딸기가 포함되어 있다면 딸기가 포함됨 출력 / 없다면 딸기가 없음

if "딸기" in fruits:
    print("딸기가 포함됨")
else:
    print("딸기가 없음")
```
